**Title -**
Develop a sentiment analysis model to analyze customer reviews of products or services.


```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("swiggy1.csv")
df2 = pd.read_csv("zomato1.csv")
```

```python
def label_sentiment(rating):
    if rating >= 4:
        return "Positive"
    elif rating <= 2:
        return "Negative"
    else:
        return "Neutral"
```

```python
def sentiment_accuracy(df, name="Dataset"):

    from sklearn.model_selection import train_test_split
    from sklearn.feature_extraction.text import TfidfVectorizer
    from sklearn.linear_model import LogisticRegression
    from sklearn.metrics import accuracy_score, classification_report

    print("\nProcessing:", name)

    df = df[['review_description', 'rating']].copy()
    df = df.dropna()

    print("Number of samples:", len(df))

    df['sentiment'] = df['rating'].apply(label_sentiment)

    X_train, X_test, y_train, y_test = train_test_split(
        df['review_description'],
        df['sentiment'],
        test_size=0.2,
        random_state=42
    )

    tfidf = TfidfVectorizer(stop_words='english', max_features=3000)

    X_train_tfidf = tfidf.fit_transform(X_train)
    X_test_tfidf = tfidf.transform(X_test)

    model = LogisticRegression(max_iter=2000, n_jobs=-1)
    model.fit(X_train_tfidf, y_train)

    y_pred = model.predict(X_test_tfidf)

    acc = accuracy_score(y_test, y_pred)

    print("Accuracy:", acc)
    print(classification_report(y_test, y_pred))

    return acc
```

```python
swiggy_acc = sentiment_accuracy(df1, "Swiggy")
zomato_acc = sentiment_accuracy(df2, "Zomato")

print("\nFinal Results")
print("Swiggy Accuracy:", swiggy_acc)
print("Zomato Accuracy:", zomato_acc)
```
